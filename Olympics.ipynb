```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load the datset
Athletes = pd.read_csv("athlete_events.csv")
Regions = pd.read_csv("noc_regions.csv")
```

```python
Athletes.head()
```

```python
Regions.head()
```

```python
#join the dataframes
Athletes_df = Athletes.merge(Regions, how = 'left', on = 'NOC')
Athletes_df.head()
```

```python
Athletes_df.shape
```

```python
#Column names consistent
Athletes_df.rename(columns = {'region': 'Region', 'notes':'Notes'}, inplace = True);
Athletes_df.head()
```

```python
Athletes_df.info()
```

```python
Athletes_df.describe()
```

```python
#Check For Null Values
nan_values = Athletes_df.isna()
nan_columns = nan_values.any()
nan_columns 
```

```python
Athletes_df.isnull().sum()
```

```python
#Top 10 COuntries participating
Top_10_Countries = Athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
Top_10_Countries
```

```python
#Plot for the top 10 Countries
plt.figure(figsize = (12,6))
plt.xticks(rotation =20)
plt.title('Overall Participation by Country')
sns.barplot(x=Top_10_Countries, y=Top_10_Countries, palette = 'Set2');
```

```python
#Age Distribution of the participation creating a histogram

plt.figure (figsize=(12,6))
plt.title("Age distribution of the Athletes")
plt.xlabel('Age')
plt.ylabel('Number of Participation')
plt.hist(Athletes_df.Age,bins=np.arange(10,80,4), color='pink', edgecolor='black'); 
```

```python
#Male and Female participants
gender_counts = Athletes_df.Sex.value_counts()
gender_counts
```

```python
#Pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=150, shadow=True);
```

```python
#Total medals
Athletes_df.Medal.value_counts()
```

```python
#Total number of females athletes in each olympics
female_partcipants = Athletes_df[(Athletes_df.Sex=='F') & (Athletes_df.Season =='Summer')][['Sex', 'Year']]
female_partcipants = female_partcipants.groupby('Year').count().reset_index()
female_partcipants.head()
female_partcipants.tail()
```

```python
womenOlympics = Athletes_df[(Athletes_df.Sex == 'F') & (Athletes_df.Season == 'Summer')]
```

```python
sns.set(style = "darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette="Spectral")
plt.title('Women Participants')
```

```python
menOlympics = Athletes_df[(Athletes_df.Sex == 'F') & (Athletes_df.Season == 'Summer')]
```

```python
sns.set(style = "darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=menOlympics, palette="Spectral")
plt.title('Men Participants')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
#Gold Medal Athletes
Goldmedals = Athletes_df[(Athletes_df.Medal=='Gold')]
Goldmedals.head()
```

```python
Goldmedals = Goldmedals[np.isfinite(Goldmedals['Age'])]
```

```python
#Gold beyond 60

Goldmedals['ID'][Goldmedals['Age']> 60].count()
```

```python
sports_event= Goldmedals['Sport'][Goldmedals['Age']> 60]
sports_event
```

```python
#plot for sporting event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sports_event)
plt.title('Gold Medals for Athletes over 60 years')
```

```python
#Gold Medals from each Country 
Goldmedals.Region.value_counts().reset_index(name = 'Medal').head(5)
```

```python
totalGoldMedals = Goldmedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals, 
                height=5, kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per COuntry')
```
